#W4A31

```python
chuoi = input()
hoa = 0
thuong = 0 
so = 0 
for char in chuoi:
    if chuoi.isupper():
        hoa += 1
    elif chuoi.islower():
        thuong += 1
    elif chuoi.isdigit():
        so += 1
print(hoa, thuong, so)
```
