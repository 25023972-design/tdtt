```python
def so_armstrong(x: int) -> bool:
    digits = str(abs(x))  
    n = len(digits)
    return sum(int(d) ** n for d in digits) == abs(x)
```

```python
import math

def so_chinh_phuong(x: int) -> bool:
    if x < 0:
        return False
    k = int(math.isqrt(x))
    return k * k == x
```

#W5A18

```python
def so_doi_xung(x: int) -> bool:
    if x < 0:
        return False 

    sum = x
    tong = 0

    while x > 0:
        digit = x % 10          
        tong = tong * 10 + digit  
        x //= 10                

    return sum == tong
```

#W5A19

```python
def so_ki_dieu(x: int) -> bool:
    if x < 0:
        x = -x  

    while x >= 10:
        tong = 0
        while x > 0:
            tong += x % 10   
            x //= 10
        x = tong            

    return x == 1
```

#W5A20

```python
def detect_number_type(s: str) -> str:
    s = s.strip()  

    check = s[1:] if s and s[0] in ('+', '-') else s

    if not check: 
        return 'neither'

    if check.isdigit():
        return 'int'

    parts = check.split('.')
    if len(parts) == 2 and parts[0].isdigit() and parts[1].isdigit():
        return 'float'

    return 'neither'
```

```python
def sum_halves(s: str) -> int:
    if len(s) != 6 or not s.isdigit():
        raise ValueError()

    dau  = int(s[:3])   
    cuoi = int(s[3:])   

    return dau + cuoi
```

#W5A23

```python
def quadratic_roots_count(a: float, b: float, c: float) -> int:
    if a == 0:
        if b == 0:
            return -1 if c == 0 else 0  
        return 1  

    tamgiac = b ** 2 - 4 * a * c

    if tamgiac > 0:
        return 2
    elif tamgiac == 0:
        return 1
    else:
        return 0
```

```python
def are_collinear(x1: float, y1: float,
                  x2: float, y2: float,
                  x3: float, y3: float) -> bool:

    ct = (x2 - x1) * (y3 - y1) - (x3 - x1) * (y2 - y1)

    return ct == 0
```

#W5A25

```python
def point_in_quadrilateral(quad: list, p: tuple) -> bool:
    """
    quad: [(x1,y1), (x2,y2), (x3,y3), (x4,y4)] 
    p:    (px, py) 
    """
    px, py = p
    n = len(quad) 

    def on_segment(ax, ay, bx, by):
        cross = (bx - ax) * (py - ay) - (by - ay) * (px - ax)
        if cross != 0:
            return False
        return min(ax, bx) <= px <= max(ax, bx) and \
               min(ay, by) <= py <= max(ay, by)

    for i in range(n):
        ax, ay = quad[i]
        bx, by = quad[(i + 1) % n]
        if on_segment(ax, ay, bx, by):
            return True

    count = 0
    for i in range(n):
        ax, ay = quad[i]
        bx, by = quad[(i + 1) % n]

        if (ay <= py < by) or (by <= py < ay):
            x_intersect = ax + (py - ay) * (bx - ax) / (by - ay)
            if px < x_intersect:
                count += 1

    return count % 2 == 1 
```
