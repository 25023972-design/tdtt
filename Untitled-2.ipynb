## W3A2

```python
def dao_nguoc_so_3_chu_so():
    try:
        n = int(input())
        
        if 100 <= n <= 999:
            hang_tram = n // 100
            hang_chuc = (n // 10) % 10
            hang_don_vi = n % 10
            
            so_dao_nguoc = hang_don_vi * 100 + hang_chuc * 10 + hang_tram
            
            print(f"SỐ NGUYÊN đảo ngược của {n} là: {so_dao_nguoc}")
        else:
            print("Không xử lý. Số bạn nhập không phải là số nguyên dương có đúng 3 chữ số.")
            
    except ValueError:
        print("Vui lòng nhập một số nguyên hợp lệ.")
```

## W3A3

```python
def la_luy_thua_cua_2(n):
    if n <= 0:
        return False
        
    return (n & (n - 1)) == 0

danh_sach_kiem_tra = [0, 1, 2, 4, 8, 10, 16, 18]

for so in danh_sach_kiem_tra:
    ket_qua = la_luy_thua_cua_2(so)
    print(f"{so:2} có phải lũy thừa của 2 không? {' Có' if ket_qua else ' Không'}")

```
