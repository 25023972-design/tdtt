```python
n = int(input())
copy = n
tong = 0
while copy > 0:
    cs = copy % 10
    tong += cs
    copy //= 10
print(f"tổng các chữ số của {n} = {tong}")
```

```python
n = int(input())
chan = 0 
le = 0
copy = n
while copy > 0:
    cs = copy % 10
    if cs % 2 == 0:
        chan += 1
    else:
        le += 1
    copy //= 10
print(f"{n} có {chan} cso chẵn và {le} cso lẻ")
```

```python
a = int(input())
if a == -1:
    print("k thể so sánh ")
else:
    max_a = a
    min_a = a
    while True:
        b = int(input())
        if b == -1:
            break
        elif b > max_a:
            max_a = b
        elif b < min_a:
            min_a = b
    print(f"{min_a},{max_a}")
```

```python
A = int(input("Nhập số nguyên dương A: "))
f1 = 1
f2 = 1
if A < 1:
    print("Không có số Fibonacci nào thỏa mãn.")
else:
    dapan = 1 
    while True:
        f_next = f1 + f2 
        if f_next > A:
            dapan = f2 
            break
        f1 = f2
        f2 = f_next
    print(f"Số Fibonacci lớn nhất không vượt quá {A} là: {dapan}")
```

```python
n = input("chỉ nhập chữ và khoảng trắng")
sotu = 0
dag_o_trong_tu = False
for i, char in enumerate(n):
    if char != " ":
        if not dag_o_trong_tu:
            sotu += 1
            dag_o_trong_tu = True
    else:
        dag_o_trong_tu = False
print(f"số từ  trong chuỗi là {sotu}")
```

```python
a = input("nhập 1 chuỗi")
tudautien = " "
for i, char in enumerate(a):
    if char != " ":
        tudautien += char
    else:
        if tudautien != " ":
            break
print(f"từ đầu tiên là {tudautien}")
```

```python
s = input("Nhập chuỗi: ")
chuoi_moi = ""
for char in s:
    if char.islower():
        chuoi_moi += char.upper() 
    elif char.isupper():
        chuoi_moi += char.lower() 
    else:
        chuoi_moi += char 
print(f"Chuỗi sau khi đổi: {chuoi_moi}")
```

```python
s = input("Nhập chuỗi: ")
in_hoa = 0
in_thuong = 0
chữ_so = 0
for char in s:
    if char.isupper():
        in_hoa += 1
    elif char.islower():
        in_thuong += 1
    elif char.isdigit():
        chữ_so += 1
print(f"Ký tự in hoa: {in_hoa}")
print(f"Ký tự in thường: {in_thuong}")
print(f"Ký tự số: {chữ_so}")
```
